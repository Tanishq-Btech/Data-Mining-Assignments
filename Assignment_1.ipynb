```python
import pandas as pd        #import pandas for creating dataframe for dataset
```

```python
df=pd.read_csv('/content/employee_data.csv')
df
```

```python
df.isnull().sum()        #counting missing values
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())                        #handling missing values using statistical methods
df['Experience']=df['Experience'].fillna(df['Experience'].mean())
df['Salary']=df['Salary'].fillna(df['Salary'].mean())
```

```python
df.isnull().sum()        #missing values filled
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df
```

```python
from sklearn.preprocessing import MinMaxScaler
le=MinMaxScaler()
df['Age']=le.fit_transform(df[['Age']])
df['Experience']=le.fit_transform(df[['Experience']])
df['Salary']=le.fit_transform(df[['Salary']])
```

```python
df
```

```python
mean_age=df['Age'].mean()
mode_age=df['Age'].mode().iloc[0]
median_age=df['Age'].median()
print(mean_age)
```
